```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
daxpy = np.genfromtxt("daxpy_results.csv", delimiter=',')
dgemv = np.genfromtxt("dgemv_results.csv", delimiter=',')
dgemm = np.genfromtxt("dgemm_results.csv", delimiter=',')
i = np.arange(len(daxpy)) + 2
```

```python
fig, ax = plt.subplots()
plt.xlabel("Problem Size (n)")
plt.ylabel("Average Time (sec)")
plt.title("DAXPY")
plt.plot(i, daxpy)
plt.show()

fig, ax = plt.subplots()
plt.xlabel("Problem Size (n)")
plt.ylabel("Average Time (sec)")
plt.title("DGEMV")
plt.plot(i, dgemv)
plt.show()

fig, ax = plt.subplots()
plt.xlabel("Problem Size (n)")
plt.ylabel("Average Time (sec)")
plt.title("DGEMM")
plt.plot(i, dgemm)
plt.show()
```
