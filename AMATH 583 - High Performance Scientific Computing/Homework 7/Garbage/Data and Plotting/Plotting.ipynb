```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Problem 1.
t_daxpy = np.genfromtxt("cblas_daxpy_time.csv", delimiter=',')
t_dgemv = np.genfromtxt("cblas_dgemv_time.csv", delimiter=',')
t_dgemm = np.genfromtxt("cblas_dgemm_time.csv", delimiter=',')
n = 2 * (np.arange(2048) + 1)
flops_daxpy = 2 * n / t_daxpy
flops_dgemv = 2 * (n ** 2) / t_dgemv
flops_dgemm = 2 * (n ** 3) / t_dgemm


fig, ax = plt.subplots()
plt.plot(n, flops_daxpy, label="cblas_daxpy")
plt.plot(n, flops_dgemv, label="cblas_dgemv")
plt.plot(n, flops_dgemm, label="cblas_dgemm")
plt.xlabel("Problem Size (n)")
plt.yscale('log')
plt.xscale('log')
plt.ylabel("FLOPs")
plt.title("Problem 1")
plt.legend()
plt.show()
```

```python
# Problem 2.
res = np.genfromtxt("res.csv", delimiter=',')
errs = np.genfromtxt("errs.csv", delimiter=',')
n = (2 ** (np.arange(10) + 4))

fig, ax = plt.subplots()
plt.plot(n, res, label="Residuals")
plt.plot(n, errs, label="Norm Errors")
plt.xlabel("Problem Size (n)")
plt.yscale('log')
plt.xscale('log')
plt.ylabel("Values")
plt.title("Problem 2")
plt.legend()
plt.show()
```

```python
# Problem 3.
data3 = np.genfromtxt("openblas_vs_cublas.csv", delimiter=',')
n_vals = data3[1:, 0]
flops_openblas = data3[1:, 1]
flops_cublas = data3[1:, 2]

print(n_vals.shape)
print(flops_openblas.shape)

fig, ax = plt.subplots()
plt.plot(n_vals, flops_openblas, label="Flops Openblas")
plt.plot(n_vals, flops_cublas, label="Flops Cublas")
plt.xlabel("Problem Size (n)")
plt.yscale('log')
plt.xscale('log')
plt.ylabel("Values")
plt.title("Problem 3")
plt.legend()
plt.show()
```

```python
# Problem 4.
df = pd.read_csv("gpu_copy_bandwidth.csv")

# Plot.
plt.figure(figsize=(8, 6))
plt.loglog(df["bytes"], df["h2d_bandwidth"], label="Host to Device (H2D)", marker='o')
plt.loglog(df["bytes"], df["d2h_bandwidth"], label="Device to Host (D2H)", marker='s')

plt.xlabel("Buffer Size (bytes)")
plt.ylabel("Bandwidth (bytes/sec)")
plt.title("Problem 4")
plt.legend()
plt.tight_layout()
plt.show()
```
